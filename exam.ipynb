```python
israel_interest_tares = pd.read_csv('BR.csv', sep = ',')
us_interest_rates = pd.read_csv('FEDFUNDS.csv', sep = ',')
usd_to_ils = pd.read_csv('EXR.csv', sep = ',') 

df_israel_interest_tares = pd.DataFrame(israel_interest_tares)
df_israel_interest_tares['TIME_PERIOD'] = pd.to_datetime(df_israel_interest_tares['TIME_PERIOD'])
df_israel_interest_tares['Year'] = df_israel_interest_tares['TIME_PERIOD'].dt.year
df_israel_interest_tares['Month'] = df_israel_interest_tares['TIME_PERIOD'].dt.month
df_israel_interest_tares = df_israel_interest_tares[['Year', 'Month', 'OBS_VALUE']].drop_duplicates()
df_israel_interest_tares['Change_From_Last_Month'] = df_israel_interest_tares['OBS_VALUE'].diff()
df_israel_interest_tares
```

```python
df_us_interest_rates = pd.DataFrame(us_interest_rates)
df_us_interest_rates['DATE'] = pd.to_datetime(df_us_interest_rates['DATE'])
df_us_interest_rates['Year'] = df_us_interest_rates['DATE'].dt.year
df_us_interest_rates['Month'] = df_us_interest_rates['DATE'].dt.month
df_us_interest_rates = df_us_interest_rates[['Year', 'Month', 'FEDFUNDS']].drop_duplicates()
df_us_interest_rates['Change_From_Last_Month'] = df_us_interest_rates['FEDFUNDS'].diff()
df_us_interest_rates
```

```python
df_usd_to_ils = pd.DataFrame(usd_to_ils)
df_usd_to_ils['TIME_PERIOD'] = pd.to_datetime(df_usd_to_ils['TIME_PERIOD'])
df_usd_to_ils['Year'] = df_usd_to_ils['TIME_PERIOD'].dt.year
df_usd_to_ils['Month'] = df_usd_to_ils['TIME_PERIOD'].dt.month
df_usd_to_ils = df_usd_to_ils[['Year', 'Month', 'OBS_VALUE']].drop_duplicates()
df_usd_to_ils['Change_From_Last_Month'] = df_usd_to_ils['OBS_VALUE'].diff()
df_usd_to_ils
```

```python
# B - 1
# your code here
merged_interest_rates = df_israel_interest_tares.merge(df_us_interest_rates, on=['Year', 'Month'])
merged_interest_rates
```

```python
# B - 2
# your code here
merged_interest_rates['Rate_Difference'] = merged_interest_rates['OBS_VALUE'] - merged_interest_rates['FEDFUNDS']
merged_interest_rates
```

```python
# B - 3
# your code here
conditions = [
   merged_interest_rates['Rate_Difference'] >= 1,
   merged_interest_rates['Rate_Difference'] <= -1,
   (merged_interest_rates['Rate_Difference'] > -1 ) & (merged_interest_rates['Rate_Difference'] < 1 )
]
choices = ['IH', 'UH', 'N']
merged_interest_rates['Rate_Comparison'] = np.select(conditions, choices)
merged_interest_rates
```

```python
# B - 4
# your code here
merged_all = merged_interest_rates.merge(df_usd_to_ils, on=['Year', 'Month'])
merged_all
```

```python
# B - 5
# your code here
merged_all = merged_all.rename(columns={
    'OBS_VALUE_x': 'Israel_Rate',
    'Change_From_Last_Month_x': 'Israel_Change_From_Last_Month',
    'FEDFUNDS': 'US_Rate',
    'Change_From_Last_Month_y': 'US_Change_From_Last_Month',
    'OBS_VALUE_y': 'Exchange_Rate',
    'Change_From_Last_Month': 'Exchange_Change_From_Last_Month'
})
merged_all

correlation = merged_all[['Rate_Difference', 'Exchange_Rate']].corr()
correlation
```
